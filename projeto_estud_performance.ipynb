```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance1.csv')
```

```python
df
```

```python
 df.shape
```

```python
df.rename(columns={
    'gender': 'genero',
    'race/ethnicity':'raca',
    'parental level of education': 'escolaridade_parental',
    'lunch': 'almoço',
    'test preparation course': 'teste_preparatorio',
    'math score': 'nota_mat',
    'reading score': 'nota_leitura',
    'writing score': 'nota_redacao'
}, inplace=True)
```

```python
df.head()
```

```python
#campos nulos
nulos = df.isnull()

plt.figure( figsize=(16,5 ) )
plt.title('Análise de campos nulos')
sns.heatmap( nulos, cbar=False)
```

```python
#mapeando linhas
mapeamento_genero = {
    'female': 'feminino',
    'male': 'masculino'
}
df['genero'] = df['genero'].map(mapeamento_genero)
```

```python
df['genero'].head()
```

```python
mapeamento_escolaridade_parental = {
    'some college': 'alguma faculdade',
    "associate's degree": 'diplomado',
    'high school': 'ensino_medio',
    'some high school': 'escola_secundaria',
    "bachelor's degree": 'bacharel',
    "master's degree": 'mestrado'
}

df['escolaridade_parental'] = df['escolaridade_parental'].map(mapeamento_escolaridade_parental)
```

```python
df['escolaridade_parental'].head()
```

```python
mapeamento_almoco = {
    'standard': 'padrao',
    'free/reduced': 'gratis/reduzido'
}
df['almoço'] = df['almoço'].map(mapeamento_almoco)
```

```python
df['almoço'].head()
```

```python
mapeamento_teste_preparatio = {
    'none': 'nulo',
    'completed': 'completado'
}
df['teste_preparatorio'] = df['teste_preparatorio'].map(mapeamento_teste_preparatio)
```

```python
df['teste_preparatorio'].head()
```
